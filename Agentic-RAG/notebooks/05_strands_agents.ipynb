# 🤗 Agentic RAG with smolagents

Smolagents (by Hugging Face) is a highly minimal, code-centric agent framework. We use the `CodeAgent` which writes python code to solve tasks.

```
import os
os.environ['GEMINI_API_KEY'] = os.environ.get('GEMINI_API_KEY', 'YOUR_API_KEY')
```

```
from smolagents import CodeAgent, LiteLLMModel, tool
from pypdf import PdfReader

# Configure the Gemini model via LiteLLM
model = LiteLLMModel('gemini/gemini-2.5-flash')

@tool
def extract_pdf_text(filepath: str, start_page: int, num_pages: int) -> str:
    """
    Extracts text from a PDF file.
    Args:
        filepath: The path to the PDF file (e.g. '../data/Apple_Environmental_Progress_Report_2025.pdf')
        start_page: The starting page index (0-based)
        num_pages: The number of pages to read
    """
    reader = PdfReader(filepath)
    text = ''
    end_page = min(start_page + num_pages, len(reader.pages))
    for i in range(start_page, end_page):
        text += reader.pages[i].extract_text() + '\n'
    return text
```

```
agent = CodeAgent(tools=[extract_pdf_text], model=model, add_base_tools=True)

# Ask the agent a question that requires reading the report
agent.run('Use the extract_pdf_text tool to read the first 5 pages of ../data/Apple_Environmental_Progress_Report_2025.pdf and tell me what the main topic is.')
```
