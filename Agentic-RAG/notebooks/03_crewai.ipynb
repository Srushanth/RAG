# 🚣 Agentic RAG with CrewAI

CrewAI focuses on multi-agent collaboration, allowing you to define 'roles' and 'goals' for each agent.

```
import os
os.environ['GEMINI_API_KEY'] = os.environ.get('GEMINI_API_KEY', 'YOUR_API_KEY')
os.environ['GOOGLE_API_KEY'] = os.environ.get('GEMINI_API_KEY', 'YOUR_API_KEY')
```

```
from crewai import Agent, Task, Crew, Process, LLM
from crewai_tools import PDFSearchTool

# Configure Gemini LLM
llm = LLM(model='gemini/gemini-2.5-flash')

# Initialize the PDFSearchTool
pdf_tool = PDFSearchTool(pdf='../data/Apple_Environmental_Progress_Report_2025.pdf')
```

```
researcher = Agent(
    role='Environmental Report Researcher',
    goal='Extract accurate data regarding environmental initiatives from the provided report.',
    backstory='You are an expert analyst who reads corporate environmental reports to extract key metrics and goals.',
    tools=[pdf_tool],
    llm=llm,
    verbose=True
)

summarizer = Agent(
    role='Executive Summarizer',
    goal='Condense the researcher\'s findings into a concise, actionable summary.',
    backstory='You are a skilled executive assistant who summarizes complex technical data into easy-to-read reports.',
    llm=llm,
    verbose=True
)
```

```
task1 = Task(
    description='Search the Apple Environmental Report to find all goals related to 2030.',
    expected_output='A raw list of goals and metrics for 2030 found in the report.',
    agent=researcher
)

task2 = Task(
    description='Take the raw list of goals and write a 3-bullet point executive summary.',
    expected_output='A 3-bullet point markdown summary of Apple\'s 2030 goals.',
    agent=summarizer
)
```

```
crew = Crew(
    agents=[researcher, summarizer],
    tasks=[task1, task2],
    process=Process.sequential,
    verbose=True
)

result = crew.kickoff()
print('######################\nFINAL RESULT\n######################')
print(result)
```
